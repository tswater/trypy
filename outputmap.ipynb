```python
import rasterio
import numpy as np 
```

```python
# xylem
file = '/shared_space/trypy/xylem.tif'
fp = rasterio.open(file)
xylem = fp.read(1)
# xylem = np.ma.masked_array(xylem,xylem='Nan')
print(xylem.shape)

vmin = np.nanmin(xylem)
vmax = np.nanmax(xylem)
print(vmin)
print(vmax)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(12,7)) # is this too big?
# plt.imshow(xylem, vmin=vmin, vmax=vmax)
plt.imshow(xylem,cmap=plt.get_cmap('jet'),vmin=vmin,vmax=vmax) # 'YlGn'
plt.xticks([])
plt.yticks([])
# plt.imshow(xylem)
cb = plt.colorbar(orientation='horizontal', pad=0.03, shrink = 0.8)
plt.title('Xylem Hydraulic Vulnerability Curve Perimeter with PCA',fontsize=20)
```
